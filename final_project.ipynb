```python
import pandas as pd

# Prep1: Load data
who_data = pd.read_csv('WHO-COVID-19-global-data.csv')
country_capital_data = pd.read_csv('country-capital-lat-long-population.csv')
country_population_data = pd.read_csv('country-population.csv')
countries_by_area_data = pd.read_csv('list_of_countries_by_area.csv')

# Prep2: View Data Configuration - Print the first 5 rows of each DataFrame
print("WHO-COVID-19 Global Data:")
display(who_data.head())

print("Country Capital, Latitude, Longitude, and Population Data:")
display(country_capital_data.head())

print("Country Population Data:")
display(country_population_data.head())

print("List of Countries by Area Data:")
display(countries_by_area_data.head())

# Prep3: Check for missing values in WHO-COVID-19-global-data.csv and fill them
who_data.fillna({'New_cases': 0, 'Cumulative_cases': 0, 'New_deaths': 0, 'Cumulative_deaths': 0}, inplace=True)
```

```python
# Prob1-1: Remove negative cases and deaths
initial_rows = len(who_data)
who_data = who_data[(who_data['New_cases'] >= 0) & (who_data['New_deaths'] >= 0)]
removed_rows = initial_rows - len(who_data)
print(f"Rows removed where New_cases < 0 or New_deaths < 0: {removed_rows}")
display(who_data.head())
```

```python
# Prob1-2: Remove non-existent countries
valid_countries = who_data['Country'].unique()
country_capital_data = country_capital_data[country_capital_data['Country'].isin(valid_countries)]
country_population_data = country_population_data[country_population_data['Country Name'].isin(valid_countries)]
countries_by_area_data = countries_by_area_data[countries_by_area_data['Country / Dependency'].isin(valid_countries)]
print("Data after removing non-existent countries:")
display(country_capital_data.head())
display(country_population_data.head())
display(countries_by_area_data.head())
```

```python
# Prob1-3: Synchronize country codes across datasets
country_code_mapping = who_data[['Country', 'Country_code']].drop_duplicates()
countries_by_area_data = countries_by_area_data.merge(country_code_mapping, left_on='Country / Dependency', right_on='Country', how='left', suffixes=('', '_drop'))
country_population_data = country_population_data.merge(country_code_mapping, left_on='Country Name', right_on='Country', how='left', suffixes=('', '_drop'))

# Drop duplicate columns from merges
for df in [countries_by_area_data, country_population_data]:
    df.drop([col for col in df.columns if 'drop' in col], axis=1, inplace=True)

countries_by_area_data['Alpha-2 code'] = countries_by_area_data.pop('Country_code')
country_population_data['Country Code'] = country_population_data.pop('Country_code')

print("Data after synchronizing country codes:")
display(countries_by_area_data.head())
display(country_population_data.head())
```

```python
# Prob1-4: Clear values including parentheses from area measurements
countries_by_area_data['Total in km (mi)'] = countries_by_area_data['Total in km (mi)'].str.replace(r' \(.*\)', '', regex=True)
countries_by_area_data['Land in km (mi)'] = countries_by_area_data['Land in km (mi)'].str.replace(r' \(.*\)', '', regex=True)
countries_by_area_data['Water in km (mi)'] = countries_by_area_data['Water in km (mi)'].str.replace(r' \(.*\)', '', regex=True)
print("Data after removing parentheses from area measurements:")
display(countries_by_area_data.head())
```

```python
# Prob1-5: Find countries where the annual number of New_cases in 2021 is 50,000 or more,
# while simultaneously having a population density less than 1000 persons/km²

# Filter WHO data for 2021
who_data_2021 = who_data[who_data['Date_reported'].str.contains('2021')]
annual_new_cases_2021 = who_data_2021.groupby('Country').agg({'New_cases': 'sum'}).reset_index()

# Ensure '2021' column in country_population_data is float and without commas
country_population_data['2021'] = country_population_data['2021'].replace(',', '', regex=True).astype(float)

# Ensure 'Total in km (mi)' is treated as string and clean it
countries_by_area_data['Total in km (mi)'] = countries_by_area_data['Total in km (mi)'].astype(str)
countries_by_area_data['Total in km (mi)'] = countries_by_area_data['Total in km (mi)'].str.extract('(\d+,?\d*)')[0].str.replace(',', '').astype(float)

# Calculate population density
countries_by_area_data['Population Density'] = country_population_data['2021'] / countries_by_area_data['Total in km (mi)']

# Merge datasets with a custom suffix to avoid column duplication
density_cases = pd.merge(
    annual_new_cases_2021, 
    countries_by_area_data, 
    left_on='Country', 
    right_on='Country / Dependency', 
    suffixes=('', '_drop')
)

# Drop the unnecessary '_drop' columns that result from the merge to clean up the DataFrame
density_cases.drop([col for col in density_cases.columns if 'drop' in col], axis=1, inplace=True)

density_cases = density_cases[(density_cases['New_cases'] >= 50000) & (density_cases['Population Density'] < 1000)]

display(density_cases[['Country', 'New_cases', 'Population Density']])
```

```python
# Prob1-6: Find the countries with the highest annual increase in New_cases in 2021 compared to 2020 by WHO region

# Filter WHO data for 2020
who_data_2020 = who_data[who_data['Date_reported'].str.contains('2020')]
annual_new_cases_2020 = who_data_2020.groupby('Country').agg({'New_cases': 'sum'}).reset_index()

# Calculate increase in new cases
increase_data = pd.merge(annual_new_cases_2021, annual_new_cases_2020, on='Country', suffixes=('_2021', '_2020'))
increase_data['Increase'] = increase_data['New_cases_2021'] - increase_data['New_cases_2020']

# Find the maximum increase by WHO region
increase_data = pd.merge(increase_data, who_data[['Country', 'WHO_region']].drop_duplicates(), on='Country')
max_increase_by_region = increase_data.groupby('WHO_region').agg({'Increase': 'max'}).reset_index()
max_countries_by_increase = pd.merge(max_increase_by_region, increase_data, on=['WHO_region', 'Increase'])

display(max_countries_by_increase[['WHO_region', 'Country', 'Increase']])
```

```python
# Prob1-7: Find the date with the largest decrease in weekly New_cases globally, compared to the previous day, over the entire period

who_data['Date_reported'] = pd.to_datetime(who_data['Date_reported'])
who_data.sort_values(by='Date_reported', inplace=True)

# Calculate daily changes
who_data['Daily_change'] = who_data['New_cases'].diff().fillna(0)
min_change_date = who_data[who_data['Daily_change'] == who_data['Daily_change'].min()]

display(min_change_date[['Date_reported', 'Daily_change']])
```

```python
# Prob1-8: Find the longest period during which the number of New_cases globally remained above 10,000 continuously

who_data['Above_10000'] = who_data['New_cases'] > 10000
who_data['Block'] = (who_data['Above_10000'] != who_data['Above_10000'].shift()).cumsum()
periods_above_10000 = who_data[who_data['Above_10000']].groupby('Block').agg({'Date_reported': ['min', 'max', 'count']})

longest_period = periods_above_10000['Date_reported']['count'].idxmax()
longest_period_data = periods_above_10000.loc[longest_period]

display(longest_period_data)
```

```python
# Prob1-9: Divide the population density into 10 sections and calculate the average cumulative number of confirmed cases in 2021 per section

# Using qcut to categorize data into quantiles
countries_by_area_data['Density_Quantile'] = pd.qcut(countries_by_area_data['Population Density'], 10, labels=False)

# Merge datasets with custom suffixes to resolve duplicate column issue
quantile_case_data = pd.merge(
    countries_by_area_data, 
    annual_new_cases_2021, 
    left_on='Country / Dependency', 
    right_on='Country', 
    suffixes=('_area', '_cases')
)

# Drop unnecessary columns created due to duplicate names
quantile_case_data.drop(columns=['Country_cases'], inplace=True)

# Calculate average cumulative cases per quantile
average_cases_per_quantile = quantile_case_data.groupby('Density_Quantile').agg({'New_cases': 'mean'}).reset_index()

display(average_cases_per_quantile)
```

```python
# Prob2-1

import matplotlib.pyplot as plt

# Assuming 'who_data' is already loaded and contains the 'Cumulative_cases' and 'Date_reported'
# Convert 'Date_reported' to datetime
who_data['Date_reported'] = pd.to_datetime(who_data['Date_reported'])

# Group data to get the latest cumulative cases by country
latest_cumulative_cases = who_data.groupby('Country')['Cumulative_cases'].max()

# Get the top 10 countries with the highest cumulative cases
top_10_countries = latest_cumulative_cases.nlargest(10).index

# Filter the original data to include only the top 10 countries
top_10_data = who_data[who_data['Country'].isin(top_10_countries)]

# Group by Country and resample by month to get the cumulative cases at the end of each month
monthly_cumulative = top_10_data.set_index('Date_reported').groupby('Country').resample('M').max()

# Ensuring columns don't duplicate on reset
if 'Country' in monthly_cumulative.columns:
    monthly_cumulative.drop(columns=['Country'], inplace=True)

# Reset index to flatten the DataFrame for plotting
monthly_cumulative.reset_index(inplace=True)

# Plotting
plt.figure(figsize=(14, 8))
for country in top_10_countries:
    monthly_data = monthly_cumulative[monthly_cumulative['Country'] == country]
    plt.plot(monthly_data['Date_reported'], monthly_data['Cumulative_cases'], marker='o', label=country)

plt.title('Monthly Cumulative Changes in Total Number of Infections for Top 10 Countries')
plt.xlabel('Date')
plt.ylabel('Cumulative Cases')
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# Prob2-2

# Assuming 'who_data' is already loaded and contains the 'New_deaths' and 'New_cases'
# Calculate the total new deaths and new cases by country
total_deaths_cases = who_data.groupby('Country').agg({
    'New_deaths': 'sum',
    'New_cases': 'sum'
}).reset_index()

# Avoid division by zero by replacing zeros with NaN in 'New_cases' before division
total_deaths_cases['New_cases'].replace(0, pd.NA, inplace=True)

# Calculate mortality rate as new deaths divided by new cases
total_deaths_cases['Mortality Rate'] = total_deaths_cases['New_deaths'] / total_deaths_cases['New_cases']

# Sort by the highest mortality rate and select the top 10
top_10_mortality_rates = total_deaths_cases.sort_values('Mortality Rate', ascending=False).head(10)

# Plotting
plt.figure(figsize=(12, 8))
plt.bar(top_10_mortality_rates['Country'], top_10_mortality_rates['Mortality Rate'], color='maroon')
plt.title('Top 10 Countries with the Highest Mortality Rates')
plt.xlabel('Country')
plt.ylabel('Mortality Rate (Deaths/Cases)')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to make room for the rotated x-axis labels
plt.grid(True, axis='y')

plt.show()
```

```python
# Prob2-3

# Ensure exact country name in the dataset
country_name = "Congo"  # Use the correct name found from the dataset

# Filter data for the AFRO region
afro_data = who_data[who_data['WHO_region'] == 'AFRO']

# Pivot table for cumulative cases by country over time
cumulative_cases_by_country = afro_data.pivot_table(
    values='Cumulative_cases', 
    index='Date_reported', 
    columns='Country', 
    aggfunc='max'
).ffill().fillna(0)  # Using forward fill and then filling remaining NaNs with 0

# Validate the presence of the country in the data
if country_name not in cumulative_cases_by_country.columns:
    raise ValueError(f"{country_name} data is not available in the dataset.")

# Calculate correlation coefficients with the target country
drc_cases = cumulative_cases_by_country[country_name]
correlations = {}
for country in cumulative_cases_by_country.columns:
    if country != country_name:
        country_cases = cumulative_cases_by_country[country]
        if country_cases.var() == 0 or drc_cases.var() == 0:
            correlations[country] = np.nan
        else:
            correlations[country] = drc_cases.corr(country_cases)

# Filter out countries with NaN correlations
correlations = {k: v for k, v in correlations.items() if not np.isnan(v)}

# Sort countries by correlation coefficients
sorted_correlations = sorted(correlations.items(), key=lambda x: x[1], reverse=True)

# Select the nearest and furthest countries
nearest_countries = [country for country, _ in sorted_correlations[:4]]
furthest_countries = [country for country, _ in sorted_correlations[-4:]]

# Combine selected countries
selected_countries = nearest_countries + furthest_countries

# Create a dictionary of the selected correlations
selected_correlations = {country: correlations[country] for country in selected_countries}

# Plotting
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12, 8))
fig.subplots_adjust(hspace=0.1)

# Define colors for the bars
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c4e17f', '#76d7c4']

# Top plot (0.9 to 1.0)
ax1.bar(selected_correlations.keys(), selected_correlations.values(), color=colors)
ax1.set_ylim(0.9, 1.0)
ax1.set_yticks(np.arange(0.9, 1.1, 0.05))
ax1.grid(True, axis='y', linestyle='--', linewidth=0.5)
ax1.spines['bottom'].set_visible(False)

# Bottom plot (0.0 to 0.4)
ax2.bar(selected_correlations.keys(), selected_correlations.values(), color=colors)
ax2.set_ylim(0.0, 0.4)
ax2.set_yticks(np.arange(0.0, 0.5, 0.1))
ax2.grid(True, axis='y', linestyle='--', linewidth=0.5)
ax2.spines['top'].set_visible(False)

# Hide the spines between ax1 and ax2
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.tick_params(labeltop=False)  # don't put tick labels at the top
ax2.xaxis.tick_bottom()

# Diagonal lines to indicate the break in the y-axis
d = .015  # how big to make the diagonal lines in axes coordinates
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)  # top-left diagonal
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

# Labels and titles
fig.suptitle('Correlation of Cumulative COVID-19 Cases with Congo', fontsize=16)
ax2.set_xlabel('Country', fontsize=12)
ax1.set_ylabel('Correlation Coefficient', fontsize=12)
ax2.set_ylabel('Correlation Coefficient', fontsize=12)
plt.xticks(rotation=45, fontsize=10)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to make room for the suptitle
plt.show()
```
