```python
import pandas as pd
from IPython.display import display  # Import the display function for better visualization in Jupyter Notebook


# Prep
# Load the necessary datasets (Prep1: Load data)
who_data = pd.read_csv('WHO-COVID-19-global-data.csv')
country_capital_data = pd.read_csv('country-capital-lat-long-population.csv')
country_population_data = pd.read_csv('country-population.csv')
countries_by_area_data = pd.read_csv('list_of_countries_by_area.csv')

# Prep2: View Data Configuration - Display the first 5 rows of each dataframe
print("First 5 rows of WHO-COVID-19 global data:")
display(who_data.head())

print("First 5 rows of Country-Capital-Lat-Long-Population data:")
display(country_capital_data.head())

print("First 5 rows of Country Population data:")
display(country_population_data.head())

print("First 5 rows of List of Countries by Area data:")
display(countries_by_area_data.head())

# Prep3: Check for missing values in WHO-COVID-19-global-data.csv and fill them
# Numerical columns are filled with zeros where needed; non-numerical columns are not modified
who_data.fillna({
    'New_cases': 0, 'Cumulative_cases': 0, 
    'New_deaths': 0, 'Cumulative_deaths': 0
}, inplace=True)

# Display the first 5 rows of WHO data after filling missing values
print("First 5 rows of WHO-COVID-19 global data after filling missing values:")
display(who_data.head())
```

```python
# Prob1-1

# Load the WHO-COVID-19 global data
who_data = pd.read_csv('WHO-COVID-19-global-data.csv')

# Remove rows where 'New_cases' < 0
who_data = who_data[who_data['New_cases'] >= 0]

# Remove rows where 'New_deaths' < 0
who_data = who_data[who_data['New_deaths'] >= 0]

# Display the modified data to verify that rows are removed
print("Data after removing rows with negative New_cases and New_deaths:")
display(who_data.head())
```

```python
# Prob1-2

# Extract unique countries from the WHO data
unique_countries = who_data['Country'].unique()

# Filter the other datasets to include only rows with countries that exist in the WHO data
filtered_country_capital_data = country_capital_data[country_capital_data['Country'].isin(unique_countries)]
filtered_country_population_data = country_population_data[country_population_data['Country Name'].isin(unique_countries)]
filtered_countries_by_area_data = countries_by_area_data[countries_by_area_data['Country / Dependency'].isin(unique_countries)]

# Display the filtered data (optional, for verification)
print("Filtered Country Capital Data:")
display(filtered_country_capital_data.head())

print("Filtered Country Population Data:")
display(filtered_country_population_data.head())

print("Filtered Countries by Area Data:")
display(filtered_countries_by_area_data.head())
```

```python
# Prob1-3

# Create a mapping from WHO data for country codes
country_code_mapping = who_data[['Country', 'Country_code']].drop_duplicates()

# Merge and update the 'Alpha-2 code' in countries by area data
countries_by_area_data = countries_by_area_data.merge(country_code_mapping, left_on='Country / Dependency', right_on='Country', how='left')
countries_by_area_data['Alpha-2 code'] = countries_by_area_data['Country_code']
countries_by_area_data.drop(columns=['Country_code', 'Country'], inplace=True)

# Merge and update the 'Country Code' in country population data
country_population_data = country_population_data.merge(country_code_mapping, left_on='Country Name', right_on='Country', how='left')
country_population_data['Country Code'] = country_population_data['Country_code']
country_population_data.drop(columns=['Country_code', 'Country'], inplace=True)

# Display the updated data to verify the changes
print("Updated Countries by Area Data:")
display(countries_by_area_data.head())

print("Updated Country Population Data:")
display(country_population_data.head())
```
